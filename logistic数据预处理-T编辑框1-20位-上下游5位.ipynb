```python
import pandas as pd
import numpy as np
import scipy
```

```python
df_ava = pd.read_csv('1249T61-90&4行列均值总表（1%）.csv')
df_ava
```

```python
n20_list = []
eff_list = []
q = 0 #   行
k = 0
for line in df_ava['N20']:
#     k+=1
    for i in range(5,25):
        if line[i]=='T':
            if df_ava['{}_G'.format(i-4)][q]==-1:
                continue
            if df_ava['{}_C'.format(i-4)][q]>=df_ava['{}_G'.format(i-4)][q] and df_ava['{}_C'.format(i-4)][q]>=df_ava['{}_A'.format(i-4)][q]: #只看C-G就可以还是都看？    df_ava['2_A_x'][q]>0 or df_ava['2_T_x'][q]>0 or 
                n20_list.append(line[i-5:i+6])
                eff_list.append(1)
            else:
                n20_list.append(line[i-5:i+6])
                eff_list.append(0)
    q+=1
#     if k>5:
#         break
```

```python
n20_list
```

```python
eff_list
```

```python
df_lab = pd.DataFrame({'N20':n20_list,
                      'T->Aefficiency':eff_list})
df_lab
```

```python
df_acgt = pd.concat([df_lab['T->Aefficiency'], df_lab['N20'].str.split('', expand=True)], axis=1)
df_acgt
```

```python
df_acgt2 = df_acgt.drop(['T->Aefficiency'],axis=1)
df_acgt2 = df_acgt2.drop(df_acgt2.columns[[0, 12]], axis=1) 
df_acgt2 = df_acgt2.dropna()
df_acgt2
```

```python
X_df_acgt = pd.get_dummies(df_acgt2) #每一列有什么就怎么命名
X_df_acgt['T->Aefficiency'] = df_acgt['T->Aefficiency']
X_df_acgt.dropna(inplace=True)
X_df_acgt
```

```python
y = X_df_acgt['T->Aefficiency']

X_df_acgt.drop(['T->Aefficiency'],axis=1, inplace=True)
X = X_df_acgt
X
```

```python
#以L1正则化本质是一个特征选择的过程
#如果我们的主要目的只是为了防止过拟合，选择L2正则化就足够了
from sklearn.linear_model import LogisticRegression as LR
from sklearn.datasets import load_breast_cancer
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import r2_score
```

```python
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.2,random_state=420)
```

```python
lrl1 = LR(penalty='l1',solver='liblinear',C=0.5,max_iter=1000)
lrl1 = lrl1.fit(X,y)
lrl1.coef_
```

```python
accuracy_score(lrl1.predict(Xtrain),Ytrain)
```

```python
accuracy_score(lrl1.predict(Xtest),Ytest)
```

```python
lrl1.fit(X,y).coef_
```

```python
lrl2 = LR(penalty='l2',solver='liblinear',C=0.5,max_iter=1000)
lrl2 = lrl2.fit(X,y)
lrl2.coef_
```

```python
accuracy_score(lrl2.predict(Xtrain),Ytrain)
```

```python
accuracy_score(lrl2.predict(Xtest),Ytest)
```

```python
l1 = []
l2 = []
l1test = []
l2test = []
Xtrain,Xtest,Ytrain,Ytest = train_test_split(X,y,test_size=0.2,random_state=420)

for i in np.linspace(0.05,1,19):
    lrl1 = LR(penalty='l1',solver='liblinear',C=i,max_iter=1000)
    lrl2 = LR(penalty='l2',solver='liblinear',C=i,max_iter=1000)
    lrl1 = lrl1.fit(Xtrain,Ytrain)
    
    l1.append(accuracy_score(lrl1.predict(Xtrain),Ytrain))
    l1test.append(accuracy_score(lrl1.predict(Xtest),Ytest))
    lrl2 = lrl2.fit(Xtrain,Ytrain)
    
    l2.append(accuracy_score(lrl2.predict(Xtrain),Ytrain))
    l2test.append(accuracy_score(lrl2.predict(Xtest),Ytest))

graph = [l1,l2,l1test,l2test]
color = ['green','black','lightgreen','gray']
label = ['L1','L2','L1test','L2test']

plt.figure(figsize=(6,6))
for i in range(len(graph)):
    plt.plot(np.linspace(0.05,1,19),graph[i],color[i],label=label[i])
plt.legend(loc=4)
plt.show()#差异不是很大L1,L2,默认L2正则化
```
